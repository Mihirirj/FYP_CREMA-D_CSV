```python
import os
import librosa
import numpy as np
import pandas as pd

# Set dataset path import os
import librosa
import numpy as np
import pandas as pd

# Set dataset path 
dataset_path = "/content/drive/MyDrive/AudioWAV"
csv_output = "/content/audio_features.csv"  # Output CSV file

# Create a list to store extracted features
data = []

# Function to extract features from an audio file
def extract_features(file_path):
    y, sr = librosa.load(file_path, sr=None)  # Load audio

    # Extract audio features
    mfccs = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13).T, axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(y=y, sr=sr).T, axis=0)
    zcr = np.mean(librosa.feature.zero_crossing_rate(y).T, axis=0)
    spec_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr).T, axis=0)

    # Combine features
    features = np.hstack([mfccs, chroma, zcr, spec_centroid])
    return features

# Process all .wav files in the dataset folder
for file_name in os.listdir(dataset_path):
    if file_name.endswith(".wav"):  # Process only .wav files
        file_path = os.path.join(dataset_path, file_name)
        features = extract_features(file_path)
        data.append([file_name] + features.tolist())  # Append filename and features

# Define column names
mfcc_labels = [f"MFCC_{i}" for i in range(13)]
chroma_labels = [f"Chroma_{i}" for i in range(12)]
columns = ["Filename"] + mfcc_labels + chroma_labels + ["ZCR", "Spectral_Centroid"]

# Convert to DataFrame and save to CSV
df = pd.DataFrame(data, columns=columns)
df.to_csv(csv_output, index=False)

print(f"CSV file saved as {csv_output}")

dataset_path = "/content/drive/MyDrive/AudioWAV"
csv_output = "/content/audio_features.csv"  # Output CSV file

# Create a list to store extracted features
data = []

# Function to extract features from an audio file
def extract_features(file_path):
    y, sr = librosa.load(file_path, sr=None)  # Load audio

    # Extract audio features
    mfccs = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13).T, axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(y=y, sr=sr).T, axis=0)
    zcr = np.mean(librosa.feature.zero_crossing_rate(y).T, axis=0)
    spec_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr).T, axis=0)

    # Combine features
    features = np.hstack([mfccs, chroma, zcr, spec_centroid])
    return features

# Process all .wav files in the dataset folder
for file_name in os.listdir(dataset_path):
    if file_name.endswith(".wav"):  # Process only .wav files
        file_path = os.path.join(dataset_path, file_name)
        features = extract_features(file_path)
        data.append([file_name] + features.tolist())  # Append filename and features

# Define column names
mfcc_labels = [f"MFCC_{i}" for i in range(13)]
chroma_labels = [f"Chroma_{i}" for i in range(12)]
columns = ["Filename"] + mfcc_labels + chroma_labels + ["ZCR", "Spectral_Centroid"]

# Convert to DataFrame and save to CSV
df = pd.DataFrame(data, columns=columns)
df.to_csv(csv_output, index=False)

print(f"CSV file saved as {csv_output}")
```
